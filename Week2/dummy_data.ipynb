### \[Problem 1] Creating random numbers
Create 500 random numbers with a two-dimensional normal distribution whose mean is (-3, 0) and whose covariance matrix is \[1.0. 0.8], \[0.8 1.0]

Multidimensional normally distributed random numbers can be created by using np.random.multivariate_normal() .

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
```

```python
np.random.seed(0)
ndarray = np.random.multivariate_normal(mean, cov, 500)
ndarray.shape
```

### \[Problem 2] Visualization by scatter plot

```python
plt.scatter(ndarray[:,0], ndarray[:,1], c='orange') # x = dimension 1, y = dimension 2
plt.title("Random 500 number scatter plot")
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

### \[Problem 3] Visualization by histogram

```python
plt.hist(ndarray[:,0], color='red', bins=40)
plt.title('Histogram of x1')
plt.xlabel('x1')
plt.xlim(-5,2)
plt.ylabel('frequency')
plt.show()
plt.hist(ndarray[:,1], color='blue', bins=40)
plt.title('Histogram of x2')
plt.xlabel('x2')
plt.xlim(-5,2)
plt.ylabel('frequency')
plt.show()
```

### \[Problem 6] Labeling

We know that by using vstack function, the data range from 1st to 500th comes from the Problem 1 array, and the other half comes from the Problem 4 array.

Therefore, we will create the third array which have the zeroes range from 1st element to 500th element to indicate that these values are belong to Problem 1 array, and the other values are labeled with ones.

```python
zeroes = np.zeros(500)
ones = np.ones(500)
label = np.concatenate((zeroes, ones)) # Combine the zrray of 0s and 1s into the (1000,) array
label.shape
```
