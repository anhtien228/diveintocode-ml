### \[Problem 1] Linear function

### \[Problem 2] Array combination

```python
x_array = np.arange(-50, 50.1, 0.1)
y_array = np.array(0.5*x_array + 1)
combine_array = np.dstack((x_array, y_array))
print("Array shape: {}".format(combine_array.shape))
print("First 20 values \n{}".format(combine_array))
```

### \[Problem 3] Find the gradient

```python
np.set_printoptions(suppress=True)
x_gradient = x_array[1:] - x_array[:-1]
y_gradient = y_array[1:] - y_array[:-1]

print("Change in x coordinate: \n{}\n".format(x_gradient))
print("Change in y coordinate: \n{}\n".format(y_gradient))

gradient = y_gradient/x_gradient
print("Gradient: \n{}\n".format(gradient))
```

### \[Problem 4] Draw a graph

```python
plt.plot(x_array[1:], gradient, c='blue') # x = dimension 1, y = dimension 2
plt.title("Linear function graph")
plt.xlabel('x-axis')
plt.ylabel('gradient')
plt.show()
```

### \[Problem 5] Python functionalization

```python
def gradient_calculate(x, y_function):
    """
    Function:
        Calculate the gradient from x (input) and y (linear function), then plot the graph to show its change

    Argument:
    x (tuple):
        Value of the function
    y_function (function):
        List of the function
    
    Output:
    array_xy (ndarray, shape(n,2)):
        A combination of x and y. n depends on x_range.
    gradient (array):
        Change in y divide for change in x (dY/dX)
    """

    array_xy = np.dstack((x, y_function))
    x_gradient = x[1:] - x[:-1]
    y_gradient = y_function[1:] - y_function[:-1]
    gradient = y_gradient/x_gradient

    return combine_array, gradient

def function1(array_x):
    array_y = array_x**2
    return array_y

def function2(array_x):
    array_y = 2*(array_x)**2 + 2**(array_x)
    return array_y

def function3(array_x):
    array_y = np.sin(array_x**(1/2))
    return array_y
```

```python
np.set_printoptions(suppress=True)
x_function1 = np.arange(-50, 50.1, 0.1)
x_function2 = np.arange(-50, 50.1, 0.1)
x_function3 = np.arange(0, 50.1, 0.1)
```

```python
array_function1, gradient1 = gradient_calculate(x_function1, function1(x_function1))
array_function2, gradient2 = gradient_calculate(x_function2, function2(x_function2))
array_function3, gradient3 = gradient_calculate(x_function3, function3(x_function3))
```

```python
fig, axs = plt.subplots(3, 1, figsize=(5,5))
plt.figure(figsize=(6,10))
axs[0].plot(x_function1[1:], gradient1)
axs[0].set_title('Function 1 gradient')
axs[1].plot(x_function2[1:], gradient2, 'tab:red')
axs[1].set_title('Function 2 gradient')
axs[2].plot(x_function3[1:], gradient3, 'tab:pink')
axs[2].set_title('Function 3 gradient')
fig.tight_layout()
plt.show()
```

### \[Problem 6] Find the minimum value

```python
min_value1 = np.min(function1(x_function1))
min_index1 = np.argmin(function1(x_function1))
min_value2 = np.min(function1(x_function2))
min_index2 = np.argmin(function1(x_function2))
min_value3 = np.min(function1(x_function3))
min_index3 = np.argmin(function1(x_function3))

fig, axs = plt.subplots(3, 1, figsize=(5,5))
plt.figure(figsize=(6,10))
axs[0].plot(x_function1, function1(x_function1))
axs[0].plot(min_value1, min_index1, marker="o")
axs[0].set_title('Function 1 slope')
axs[1].plot(x_function2, function2(x_function2), color="red")
axs[1].plot(min_value2, min_index2, marker="o")
axs[0].set_title('Function 2 slope')
axs[2].plot(x_function3, function3(x_function3), color="cyan")
axs[2].plot(min_value3, min_index3, marker="o")
axs[0].set_title('Function 3 slope')
fig.tight_layout()
plt.show()
```
