### Problem 1
Create a code to calculate how many grains of rice you receive in total on the 100th day, and output the answer. Also, use a line graph to represent the change in the number of rice grains you receive on that day and the total number of rice grains you receive by that day from day 1 to day 100. (The vertical axis is the number of rice and the horizontal axis is the number of days)

```python
GRAIN = 1               # The grain of rice received on the first day
total_grain = GRAIN     # Accumulation of grain of rice
grain_per_day = []
grain_accumulated = []
grain_per_day.append(GRAIN)
grain_accumulated.append(GRAIN)
days = 100

# Calculate number of grains of rice on the 100th day
for i in range(1, days):
    GRAIN*=2
    total_grain+=GRAIN
    grain_per_day.append(GRAIN)
    grain_accumulated.append(total_grain)

print("The total grains of rice: {}".format(total_grain))
```

```python
print(len(grain_per_day))
```

```python
"""
Display the graph. Title and axis label name.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Grains of rice on the 100th day")
plt.xlabel("number of days ")
plt.ylabel("number of rice")
plt.plot(grain_per_day, label="grain per day", color="blue")
plt.plot(grain_accumulated, label="grain accumulated", color="red")
plt.legend()
plt.show()
```

### Problem 2
The number of days in this anecdote is ambiguous, as it could be 30 days or 81 days. Let's create a function that takes the number of days as an argument so that it can be used for any story. After creating the function, answer how many grains of rice you can get for the number of days you take the course. As in the previous question, you should also output a line graph.


The following template contains not only the framework as a function, but also comments about arguments and return values. Please refer to it and rewrite the comments to make it easier to understand for yourself as well.

```python
def compute_sorori_shinzaemon(days=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.

    Parameteres:
    --------------
    days: int
        Number of days to get rice (default : 100)

    Returns:
    --------------
    grain_per_day : list
        List of the number of rice you get one day (the length of the list is equal to day)
    grain_accumulated : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    GRAIN = 1               # The grain of rice received on the first day
    total_grain = GRAIN     # Accumulation of grain of rice
    grain_per_day = []
    grain_accumulated = []
    grain_per_day.append(GRAIN)
    grain_accumulated.append(GRAIN)

    # Calculate number of grains of rice on the 100th day
    for i in range(1, days):
        GRAIN*=2
        total_grain+=GRAIN
        grain_per_day.append(GRAIN)
        grain_accumulated.append(total_grain)

    return grain_per_day, grain_accumulated
```

```python
grain_per_day, grain_accumulated = compute_sorori_shinzaemon(200)
```

```python
# Display the graph 
plt.title("Grains of rice on the 100th day")
plt.xlabel("number of days ")
plt.ylabel("number of rice")
plt.plot(grain_per_day, label="grain per day", color="blue")
plt.plot(grain_accumulated, label="grain accumulated", color="red")
plt.legend()
plt.show()
```
