### Problem 1

```python
import time # Module used to compare execution time
```

```python
"""
Code to calculate the thickness when the paper is folded once
"""
THICKNESS = 0.00008
start = time.time()
folded_thickness = THICKNESS*(2**43)
elapsed_time = time.time() - start
print("Thickness: {} meters".format(folded_thickness))
print("Time executed: {:.10f}[ms]".format(elapsed_time)) # I modified the code to show the milliseconds for easier observeation
```

### Problem 2

```python
# Convert the thickness unit to kilometers with 2 decimal places
print("Thickness: {:.10f} kilometers".format(folded_thickness/1000))
```

### Problem 3

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS
start = time.time()
for i in range(43):
    folded_thickness = folded_thickness*2

elapsed_time = time.time() - start

print("Thickness (for loop): {} meters".format(folded_thickness))
print("Time executed (for loop): {:.10f}[s]".format(elapsed_time)) # I modified the code to show the milliseconds for easier observeation
```

### Problem 4

```python
%%timeit
THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(43):
    folded_thickness = folded_thickness*2

print("Thickness (for loop): {} meters".format(folded_thickness))
print("Thickness (for loop): {: .2f} kilometers".format(folded_thickness/1000))
```

### Problem 5

```python
data = []
THICKNESS = 0.00008
folded_thickness = THICKNESS
data.append(THICKNESS)
for i in range(43):
    folded_thickness = folded_thickness*2
    data.append(folded_thickness)

print("Length of list: {}".format(len(data)))
```

### Problem 6

```python
"""
Display the graph. Title and axis label name.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(data) # Enter the variable name of the list in "List name"
plt.show()
```

### Problem 7

```python
"""
Display a red line graph.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=12)
plt.plot(data, color='orange', linestyle='dotted', linewidth=2, markersize=12) # I prefer the orange color
plt.show()
```
